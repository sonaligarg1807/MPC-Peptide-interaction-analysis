```python
#initialization
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.interpolate import make_interp_spline

#mount gdrive
from google.colab import drive
drive.mount('/content/gdrive',force_remount=True)
```

```python
a,b=np.loadtxt('gdrive/My Drive/xvg files/rmsd_mel-none.xvg', unpack=True)

#smooth curve
X_Y_Spline = make_interp_spline(a, b)
X_ = np.linspace(a.min(), a.max(), 1000)
Y_ = X_Y_Spline(X_)

#adding label, markers that is points on curve
plt.plot(X_,Y_, color='mediumpurple')

#adding x and y label
plt.xlabel("Time(ns)", fontdict={'fontname':'Comic Sans MS'}, fontsize=16)
plt.ylabel("RMSD(nm)",fontdict={'fontname':'Comic Sans MS'}, fontsize=16)
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
plt.xlim(0,500)

#to save the file in svg format
plt.savefig('gdrive/My Drive/svg_files_analysis/RMSD of mel-mpc-graph.svg',dpi=300)
plt.show()
```

```python
a,b=np.loadtxt('gdrive/My Drive/xvg files/rmsf_mel-mpc-none.xvg', unpack=True)

#adding label, markers that is points on curve
plt.plot(a,b, color='mediumpurple',marker='.', markerfacecolor='black', markersize=10)

#adding x and y label
plt.xlabel("Residue number", fontdict={'fontname':'Comic Sans MS'}, fontsize=16)
plt.ylabel("RMSF(nm)",fontdict={'fontname':'Comic Sans MS'}, fontsize=16)
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)

#to save the file in svg format
plt.savefig('gdrive/My Drive/svg_files_analysis/RMSF of mel-mpc-graph.svg',dpi=300)
plt.show()
```

```python
n,m=np.loadtxt('gdrive/My Drive/xvg files/FINAL-MEL-LIG-DISTANCE', usecols=(0,2), unpack=True)

sns.kdeplot(m, shade='True', color="red", label='Lys21')
plt.xlabel("Distance(nm)", fontdict={'fontname':'Comic Sans MS'}, fontsize=16)
plt.ylabel("Probability",fontdict={'fontname':'Comic Sans MS'}, fontsize=16)
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
plt.legend()
plt.savefig('gdrive/My Drive/svg_files_analysis/dist-mel-mpc-lys21.svg',dpi=300)
plt.show()
```

```python
n,m=np.loadtxt('gdrive/My Drive/xvg files/hbond_mel-mpc_somedata_none.xvg', usecols=(0,1), unpack=True)

sns.kdeplot(m, shade='True', color="darkblue")

plt.xlabel("Number of H-bonds", fontdict={'fontname':'Comic Sans MS'}, fontsize=16)
plt.ylabel("Probability",fontdict={'fontname':'Comic Sans MS'}, fontsize=16)
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)

plt.savefig('gdrive/My Drive/svg_files_analysis/hbond-mel-mpc-smooth-graph_last100.svg',dpi=300)
plt.show()
```
